```python
"""
check prime number
"""

def prime(n):
    for i in range(2,int((n**(0.5))+1)):
        if n%i==0:
            return False
    return True


print(prime(502))
print(prime(2))
```

```python
"""
this is a function of fibonacci 
iteration
"""
def fib(n):
    if n==1 or n==2 :
        return 1
    elif n==0:
        return 0
    a=1
    b=1
    ans =0
    for i in range(0,n-2):
        ans = a+b
        t=b
        b=a+b
        a=t
    return ans

print(fib(2))
print(fib(10))
print(fib(4))
print(fib(5))
```

```python
"""
this is a function of fibonacci 
recurssion
"""
def fib(n):
    if n==1 or n==2 :
        return 1
    elif n==0:
        return 0
    return fib(n-1)+fib(n-2)

print(fib(2))
print(fib(10))
print(fib(4))
print(fib(5))

```

```python
""" sum upto n numbers using dp"""

import time
dp={}

def dpsum(n):
    sum=0
    ss= time.perf_counter()
    for i in reversed(range(1,n+1)):
        if(i in dp):
            sum+=dp[i]
            break
        else:
            sum+=i
    
    dp[n]=sum
    print(time.perf_counter()-ss)
    return sum

print(dpsum(29))
print(dpsum(20))
```

```python
'''lambda and filter'''
def sums(ls):
    sum=0
    for i in ls:
        sum+=i
    return sum

nums = list(range(1000))
filtered = filter(lambda x: x % 3 == 0 or x % 7 == 0, nums)
sums(list(filtered))


```
